```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
x = np.linspace(-2,2,100)
y = np.linspace(-2,2,100)

xx, yy = np.meshgrid(x, y)

plt.pcolormesh(xx, cmap='viridis')
plt.colorbar(label='X')
plt.show()

plt.pcolormesh(yy, cmap='plasma')
plt.colorbar(label='Y')
plt.show()
```

#### 1.3-1.4

```python
r = np.sqrt(xx**2 + yy**2)
theta = np.arctan2(yy,xx)

plt.figure()
plt.pcolormesh(xx,yy,r)
plt.colorbar(label='r')
plt.title('Radius')
plt.xlabel('x')
plt.ylabel('y')
plt.show

plt.figure()
plt.pcolormesh(xx,yy,theta, cmap='plasma')
plt.colorbar(label='')
plt.title('Angle')
plt.xlabel('x')
plt.ylabel('y')
plt.show
```

#### 1.5-1.7

```python
f = (np.cos(4*r)**2) + (np.cos(4*theta)**2)
plt.figure()
plt.pcolormesh(xx,yy,f)
plt.colorbar(label='f')
plt.xlabel('x')
plt.ylabel('y')
plt.show
```

```python
meanfx = np.mean(f, axis=1)
plt.figure()
plt.plot(x, meanfx)
plt.xlabel('x')
plt.ylabel('mean of f wrt to x axis')
plt.show

meanfy = np.mean(f, axis=0)
plt.figure()
plt.plot(y, meanfy)
plt.xlabel('y')
plt.ylabel('mean of f wrt to y axis')
plt.show
```
