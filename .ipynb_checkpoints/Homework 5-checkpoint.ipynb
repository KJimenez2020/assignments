```python
import numpy as np
from matplotlib import pyplot as plt
from scipy import optimize
import statsmodels.api as smod
```

### 1)

```python
test_data = np.genfromtxt('/data9/UN3400/data/brf_temp.txt', 
                          delimiter='\t')
data = test_data
```

### 2)

```python
def linfunc(x, a, b,c,d):
    y = a*x + b*x**2 + c*x**3 + d
    return y
    
p = (2 * np.pi)/365.5
def sinfunc(t, a, b,c):
    y = a * np.cos(p * t) + b * np.sin(p * t) + c
    return y

dpre = optimize.curve_fit(sinfunc, xdata = test_data[:,0], 
                             ydata = test_data[:,1])[0]
a = dpre[0]
b = dpre[1]
c = dpre[2]

dpre
```

```python
N, i = np.shape(data)
#time
t = data[:,0:1]
#data observed
dobs = data[:,1:2]
```

$G \cdot G^T$

```python
M = 2
G = np.zeros((N,M))
G[0:N,0:1] = np.ones((N,1))
G[0:N,1:2] = t
#G[0:N,2:3] = t**2
```

$G^T \cdot d^{obs} , m^{est}$

```python
#G * transpose(G)
GTG = np.matmul(G.T,G)
#transpose(G) * d_obs
GTd = np.matmul(G.T,dobs)
mest = np.linalg.lstsq(GTG,GTd,rcond=None)[0]   
mest
```

```python
dpre = np.matmul(G,mest)
#error
e = dobs - dpre
#variance
sigmad2 = np.matmul(e.T,e)/(N-M)
#Covariance
Cd = sigmad2 * np.linalg.inv(GTG)
#95% CI
sm = np.sqrt( np.diag(Cd).reshape(-1,1) )
```

```python
print("estimated standard deviation of the data")
print(np.sqrt(sigmad2))
print("estimated standard deviation of the data")
print(np.sqrt(sigmad2))
print("estimated solution of model parameters")
for k in range(M):
    print("model parameter %d: %.4f +/- %.4f" % (k, mest[k,0], 2.0*sm[k,0]) )
```

```python
result = smod.OLS(dobs,G).fit()
print(result.summary())
```
